# Group BY


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('deliveries.csv')
data1
```

```python
batsman=data1.groupby('batsman')
```

#### Q.Top 5 run getter batsman

```python
batsman['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

#### Q.Top 10 batsman with most 4s

```python
mask=data1['batsman_runs']==4
data2=data1[mask]
fours=data2.groupby('batsman')
fours.size().sort_values(ascending=False).head(10)
```
