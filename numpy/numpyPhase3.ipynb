## Advance Operation with Business Examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id,2021,2022,2023,2024]
sales_data=np.array([
    [1,150000,180000,220000,250000],
    [2,120000,140000,160000,190000],
    [3,200000,230000,260000,270000],
    [4,180000,210000,240000,270000],
    [5,160000,185000,205000,230000]
])
```

#### Slicing

```python
print(f"Sliced rows data\n{sales_data[0:3]}")
print(f"Sliced columns data\n{sales_data[:,1:]}")
```

### Suming data

```python
print(f"Vertical addition\n{np.sum(sales_data[:, 1:], axis=0)}")
print(f"Horizontal addition\n{np.sum(sales_data[:, 1:], axis=1)}")
```

### Minimum & Maximum Data

```python
#minimum sale per shop
min=np.min(sales_data[:, 1:],axis=1)
print(f"Minimum sale per shop\n {min}")

#Maximum sale per year
max=np.max(sales_data[:, 1:],axis=0)
print(f"Maximum sale per year\n {max}")

#Average Sales per shop
avg=np.mean(sales_data[:,1:],axis=1)
print(f"Average sale per shop\n{avg}")

#Cumilative sum per shop
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print(f"Cumilative sum per shop \n{cumsum}")
```

### Ploting the figure

```python
plt.figure(figsize=(6,4))
plt.plot(np.mean(cumsum[:,:],axis=0))
plt.xlabel("Years")
plt.ylabel("Sales")
plt.title("Avrage cumelative sales of all shops")
plt.grid(True)
plt.show()
```

## Arthematic operations in Numpy

```python
vector1=np.array([1,2,3,4,5])
vector2=np.array([5,6,7,8,9])

print(f"Addition is : {vector1+vector2}")

print(f"Multiplication is : {vector1*vector2}")

print(f"Dot Product : {np.dot(vector1,vector2)}")

#Angle btw two vector
#formula  angle=cos-1(dot product/magnitude multiplication)
angle=np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print(f"Angle btw : {angle} Degrees")
```

### Vectorize Function

```python
vector1=np.array([1,2,3,4,5,6,7,8])
def squre(a):
    return a*a
squredArray=np.vectorize(squre)

print(squredArray(vector1))
```
